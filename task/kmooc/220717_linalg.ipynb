```python
import numpy as np
import mathlib
```

## Q1)

```python
A = np.array([
    [1., 2., 3., 0.],
    [2., 1., -1., 4.],
    [-1., 1., 1., -1.],
])

x = mathlib.gauss(A)
print(x.round(3))
```

A1) $x_1=1, x_2=1, x_3=-1$

## Q4)

```python
A = np.array([
    [1, 2, -1],
    [3, 1, 2],
    [2, 1, -1],
])

B = np.array([
    [2, 0, 2],
    [1, 3, -2],
    [2, -3, 1],
])

ops = [' + ' for _ in range(A.shape[0]-1)]+[' ']
for i in range(A.shape[0]):
    for j in range(B.shape[1]):
        
        [print(f'{a}*{b}', end=op) for a,b,op in zip(A[i,:],B[:,j],ops)]
        print(f'= {sum([a*b for a,b in zip(A[i,:],B[:,j])])}', end='')
        print('\t', end='')
    print()
```

A4)

```python
np.matmul(A, B)
```
