```python
new_mpg = mpg.copy()
new_mpg['sumy'] = new_mpg['hwy'] + new_mpg['cty']
new_mpg.head()
```

```python
new_mpg['meany'] = new_mpg['sumy'] / 2
new_mpg.head()
```

```python
new_mpg.sort_values(['meany'], ascending=False).head(3)
```

## 분석 도전

```python
urllib.request.urlretrieve('https://raw.githubusercontent.com/youngwoos/Doit_Python/main/Data/midwest.csv','midwest.csv')
```

```python
midwest = pd.read_csv('midwest.csv')
midwest.head()
```

```python
midwest['ratekids'] = (midwest['poptotal']-midwest['popadults'])/midwest['poptotal']*100
```

```python
midwest.sort_values(['ratekids'], ascending=False).head()[['county','ratekids']]
```

```python
midwest['gradekids'] = midwest['ratekids'].apply(lambda x: 'large' if x >= 40. else 'middle' if x >= 30. else 'small')
midwest['gradekids'].value_counts()
```

```python
midwest['rateasian'] = midwest['popasian']/midwest['poptotal']*100
midwest.sort_values(['rateasian']).head(10)[['state','county','rateasian']]
```
